```python
import torch
import torch.nn.functional as F
tensor = torch.tensor([[1,2,3],[4,5,6]])
print("original tensor:")
print(tensor)
target_shape = (3, 3)
current_shape = tensor.shape
pad_left = 0
pad_right = 0
pad_top = 0
pad_bottom = 0
if current_shape[0] < target_shape[0]:
    pad_bottom = target_shape[0] - current_shape[0]
if current_shape[1] < target_shape[1]:
    pad_right = target_shape[1] - current_shape[1]
padded_tensor = F.pad(tensor,(pad_left, pad_right, pad_top, pad_bottom),mode='constant',value=0)
print("\npadded tensor:")
print(padded_tensor)
```
